#### Finding toxicogenomics chemicals and diseases with cancerous gene IDs

```python
res_df = tcga_df[['gene_id']]
res_df = res_df.drop_duplicates('gene_id')
joined_chem_df = pd.merge(res_df, toxicogenomics_chems_df, on=['gene_id'])
joined_chem_df
```

```python
counts_int_act = joined_chem_df['interaction_actions'].value_counts().to_frame()
counts_int_act['proportion'] = counts_int_act['interaction_actions'] / len(joined_chem_df)
counts_int_act
```

```python
## compare with proportions before merging
counts_int_act_prior = toxicogenomics_chems_df['interaction_actions'].value_counts().to_frame()
counts_int_act_prior['proportion'] = counts_int_act_prior['interaction_actions'] / len(toxicogenomics_chems_df)
counts_int_act_prior
```

```python
joined_dis_df = pd.merge(res_df, toxicogenomics_dis_df, on=['gene_id'])
joined_dis_df
```

```python
joined_dis_df = joined_dis_df.sort_values('inference_score', ascending=False)
joined_dis_df
```
