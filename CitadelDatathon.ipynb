#### Finding toxicogenomics chemicals and diseases with cancerous gene IDs

```python
res_df = tcga_df[['gene_id']]
res_df = res_df.drop_duplicates('gene_id')
joined_chem_df = pd.merge(res_df, toxicogenomics_chems_df, on=['gene_id'])
joined_chem_df
```

```python
counts_int_act = joined_chem_df['interaction_actions'].value_counts().to_frame()
counts_int_act['proportion'] = counts_int_act['interaction_actions'] / len(joined_chem_df)
counts_int_act
```

```python
## compare with proportions before merging
counts_int_act_prior = toxicogenomics_chems_df['interaction_actions'].value_counts().to_frame()
counts_int_act_prior['proportion'] = counts_int_act_prior['interaction_actions'] / len(toxicogenomics_chems_df)
counts_int_act_prior
```

```python
joined_dis_df = pd.merge(res_df, toxicogenomics_dis_df, on=['gene_id'])
joined_dis_df
```

```python
joined_dis_df = joined_dis_df.sort_values('inference_score', ascending=False)
joined_dis_df
```

```python
disease_name_counts_df = joined_dis_df.groupby('disease_name')['gene_id'].nunique()
disease_name_counts_df
```

```python
selection_df = joined_chem_df[joined_chem_df['gene_id'] == 'ENSG00000100985']
selection_df
```

```python
ineraction_counts_df = selection_df.groupby('interaction_actions').count()
ineraction_counts_df = ineraction_counts_df[['gene_id']]

ineraction_counts_df.rename(columns={'gene_id' : 'counts'}, inplace=True)
ineraction_counts_df
```

```python
ineraction_counts_df.sort_values('counts', ascending=False)
```

### getting the chemicals where the chemical name is in the diseases table

```python
just_chem_name_df = joined_dis_df[['inference_chemical_name']]
just_chem_name_df = just_chem_name_df.drop_duplicates('inference_chemical_name')
just_chem_name_df = just_chem_name_df.dropna()
just_chem_name_df.head(10)
```

```python
just_chem_name_df.rename(columns={'inference_chemical_name' : 'chemical_name'}, inplace=True)
subset_joined_chem_df = pd.merge(joined_chem_df, just_chem_name_df, on=['chemical_name'])
subset_joined_chem_df
```

```python
cts_df = subset_joined_chem_df.groupby('gene_id').count()
cts_df.rename(columns={'chemical_name' : 'counts'}, inplace=True)
cts_df = cts_df[['counts']]
cts_df = cts_df.sort_values('counts', ascending=False)
cts_df
```

```python
## ENSG00000100985 is the most common gene for the chemicals that are found in the disease table (w/ cancer genes)
subset_joined_chem_spec_gene = subset_joined_chem_df[subset_joined_chem_df['gene_id'] == 'ENSG00000100985']
subset_joined_chem_spec_gene.head(10)
```

```python
cts_sub_df = subset_joined_chem_spec_gene.groupby('interaction_actions').count()
cts_sub_df.rename(columns={'gene_id' : 'counts'}, inplace=True)
cts_sub_df = cts_sub_df[['counts']]
cts_sub_df = cts_sub_df.sort_values('counts', ascending=False)
cts_sub_df
```
